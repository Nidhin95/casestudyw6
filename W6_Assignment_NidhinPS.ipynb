# 1. Find out how many males and females participated in the test.

```python
Male=data[data['gender']=='male']
m = Male["gender"].count()
print('Number of males Participated =',m)
Female=data[data['gender']=='female']
f=Female['gender'].count()
print('Number of Females Participated =',f)
```

# 

```python
plt.figure(figsize=(8,8))
plt.hist(Male['gender'],rwidth=1,color='orange')
plt.hist(Female['gender'],rwidth=1)
plt.ylabel('Number Of Students',fontsize=15)
plt.xlabel('Gender',fontsize=15)
plt.title('Male vs Female',fontsize=20)
#adding legend manually
plt.legend(('male','female'),loc=[1.02,.9])
#plt.grid(color='red',linewidth=1.05,axis='y',alpha=2)
plt.grid(True)
plt.tick_params(grid_color='black',grid_alpha=1)
```

```python
Insights:
    1. The number of females is greater than males.
  
```

# 3. Who scores the most on average for math, reading and writing based on

● Gender
● Test preparation course

```python
data.mean()
```

```python
female_math_mean=Female['math score'].mean()
male_math_mean=Male['math score'].mean()
female_read_mean=Female['reading score'].mean()
male_read_mean=Male['reading score'].mean()
female_write_mean=Female['writing score'].mean()
male_write_mean=Male['writing score'].mean()

print('Average Male Score:\n\nMaths:',male_math_mean,'\nReading:',male_read_mean,'\nWriting:',male_write_mean)
print('\nAverage Female Score:\n\nMaths:',female_math_mean,'\nReading:',female_read_mean,'\nWriting:',female_write_mean)
```

```python
dn= data[(data['test preparation course']=='none')]
dc= data[(data['test preparation course']=='completed')]

n1_math_mean=dn['math score'].mean()
c1_math_mean=dc['math score'].mean()
n2_read_mean=dn['reading score'].mean()
c2_read_mean=dc['reading score'].mean()
n3_write_mean=dn['writing score'].mean()
c3_write_mean=dc['writing score'].mean()

print('\n\n\nAverage in terms of test preparation course [Not Done]:\n\nMaths:',n1_math_mean,'\nReading:',n2_read_mean,'\nWriting:',n3_write_mean)
print('\nAverage in terms of test preparation course [completed]:\n\nMaths:',c1_math_mean,'\nReading:',c2_read_mean,'\nWriting:',c3_write_mean)
```

# 4. What do you think about the scoring variation for math, reading and writing
based on ● Gender ● Test preparation course

```python
fem_math_std=Female['math score'].std()
mal_math_std=Male['math score'].std()
fem_read_std=Female['reading score'].std()
mal_read_std=Male['reading score'].std()
fem_write_std=Female['writing score'].std()
mal_write_std=Male['writing score'].std()

print('Deviation of Male Scores:\n\nMaths:',mal_math_std,'\nReading:',mal_read_std,'\nWriting:',mal_write_std)
print('\nDeviation of Female Scores:\n\nMaths:',fem_math_std,'\nReading:',fem_read_std,'\nWriting:',fem_write_std)


non_math_std=dn['math score'].std()
com_math_std=dc['math score'].std()
non_read_std=dn['reading score'].std()
com_read_std=dc['reading score'].std()
non_write_std=dn['writing score'].std()
com_write_std=dc['writing score'].std()

print('\n\n\nDeviation for test preparation course not done Scores:\n\nMaths:',non_math_std,'\nReading:',non_read_std,'\nWriting:',non_write_std)
print('\nDeviation for test preparation course completed Scores:\n\nMaths:',com_math_std,'\nReading:',com_read_std,'\nWriting:',com_write_std)
```
